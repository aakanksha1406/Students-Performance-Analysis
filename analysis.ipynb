```python

df['pass_reading']=np.where(df['reading score']>=40, 'p', 'f')
print("cutoff", df['pass_reading'].value_counts())

df['pass_writing']=np.where(df['writing score']>=40, 'p', 'f')
print("cutoff", df['pass_writing'].value_counts())
```

```python
df['total']=df['math score']+df['reading score']+df['writing score']
df['total'].head(5)
```

```python
df['percent']=(df['total']/3)
plt.figure(figsize=(4,4))
sns.kdeplot(data=df, hue='gender', x='percent', palette='dark:#5A9_r')
plt.show()
```

```python
df['percent'].describe()
```

```python
df['status']=df.apply(lambda x: 'f' if x['pass_maths'] == 'f' or 
                           x['pass_reading'] == 'f' or x['pass_writing'] == 'f'
                           else 'pass', axis = 1)
```

```python
df['status'].sample(10)
```

```python
# Assigning grades to the grades according to the following criteria :
# 0  - 40 marks : grade E
# 41 - 60 marks : grade D
# 60 - 70 marks : grade C
# 70 - 80 marks : grade B
# 80 - 90 marks : grade A
# 90 - 100 marks : grade O

def getgrade(percentage, status):
  if status == 'Fail':
    return 'E'
  if(percentage >= 90):
    return 'O'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  if(percentage >= 40):
    return 'D'
  else:
    return 'E'
df['grades'] = df.apply(lambda x: getgrade(x['percent'], x['status']), axis = 1 )

df['grades'].value_counts()
```

```python
fig, ax = plt.subplots()
ax.pie(df['grades'].value_counts(), labels=('B','C','D','A','O','E'), autopct='%1.1f%%', startangle=90)
```

```python
x=df.iloc[:, :-1]
y=df.iloc[:, -1]
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state=42)
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

## Model Training

#### Logistic Reg

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
accuracy=r2_score(y_test, y_pred)
accuracy
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

# creating a model
rf = RandomForestRegressor()

# feeding the training data to the model
rf.fit(x_train, y_train)

# predicting the x-test results
y_pred1 = rf.predict(x_test)

acc_rf=r2_score(y_test, y_pred1)
acc_rf
# overfit
```

```python
from sklearn.model_selection import cross_val_score

# computing accuracies for 20 different model's accuracy
accuracies = cross_val_score(estimator = rf, X = x_train, y = y_train, cv = 15)
print(accuracies)

# computing mean of accuaracies obtained by all the models
print("mean accuracy :", accuracies.mean())

# computing the standard variance of the models
print("mean standard variance : \n", accuracies.std())

accuracies
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
# feeding the training data to the model
dt.fit(x_train, y_train)

# predicting the x-test results
y_pred2 = dt.predict(x_test)
acc_dt=r2_score(y_test, y_pred2)
acc_dt
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': [3, 5, 7, None], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4], 'max_features': ['auto', 'sqrt', 'log2', None]}

grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(x_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)


```
